```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

### Answer to the Question no. 1

```python
ages_array = DF['age'].values
salaries_array = DF['salary'].values

print("Ages Array:")
print(ages_array)
print("Dtype:", ages_array.dtype)
print("Ndim:", ages_array.ndim)
print("Shape:", ages_array.shape)
print("Size:", ages_array.size)
print("\nSalaries Array:")
print(salaries_array)
print("Dtype:", salaries_array.dtype)
print("Ndim:", salaries_array.ndim)
print("Shape:", salaries_array.shape)
print("Size:", salaries_array.size)
```

### Answer to the Question no. 2

```python
highest_salary = np.max(salaries_array)
lowest_salary = np.min(salaries_array)
average_salary = np.mean(salaries_array)
average_age = np.mean(ages_array)

print(f"Highest Salary: {highest_salary}")
print(f"Lowest Salary: {lowest_salary}")
print(f"Average Salary: {average_salary}")
print(f"Average Age: {average_age}")
```

### Answer to the Question no. 3

```python
ages_greater_than_30 = ages_array[ages_array > 30]
print("Ages greater than 30:")
print(ages_greater_than_30)

count_greater_than_30 = len(ages_greater_than_30)
print(f"\nNumber of employees older than 30: {count_greater_than_30}")
```

### Answer to the Question no. 4

```python
print(ages_array) #orginal array
updated_ages_array = ages_array + 5
print("Updated Ages Array:")
print(updated_ages_array)
```

### Answer to the Question no. 5

```python
total_salary_expense = np.sum(salaries_array)
salary_range = np.max(salaries_array) - np.min(salaries_array)

print(f"Total Salary Expense: {total_salary_expense}")
print(f"Difference between maximum and minimum salary: {salary_range}")
```

### Answer to the Question no. 8

```python
it_employees = employees_df[employees_df['dept'] == 'IT']
print("Employees in the IT department:")
display(it_employees)

num_it_employees = len(it_employees)
print(f"\nTotal number of IT employees: {num_it_employees}")
```

### Answer to the Question no. 9

```python
sorted_by_salary = employees_df.sort_values(by='salary', ascending=False)
top_3_employees = sorted_by_salary.head(3)

print("Top 3 highest-paid employees (including department and age):")
display(top_3_employees[['id', 'age', 'salary', 'dept']])
```
