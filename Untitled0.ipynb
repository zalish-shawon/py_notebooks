```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values
x = np.linspace(-10, 2, 400)

# Define the two equations
y1 = 2*x + 3         # Line 1: y = 2x + 3
y2 = x - 3           # Line 2: y = x - 3

# Create the plot
plt.figure(figsize=(8, 6))

# Plot both lines
plt.plot(x, y1, label='y = 2x + 3', color='blue')
plt.plot(x, y2, label='y = x - 3', color='red')

# Mark the intersection point
plt.plot(-6, -9, 'ko')  # black dot at intersection
plt.text(-6.5, -8.5, '(-6, -9)', fontsize=12)

# Axes and grid
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.title('Graph of y = 2x + 3 and y = x - 3')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10, 2)
plt.ylim(-12, 5)

plt.show()
```
