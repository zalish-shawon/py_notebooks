### Simple 2D Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple 2D Line Plot")

# Show the plot
plt.grid(True)
```

### Creating Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0) # for reproducibility
x = np.random.rand(50)
y = np.random.rand(50)

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Scatter Plot")

# Show the plot
plt.grid(True)
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (e.g., random numbers from a normal distribution)
data = np.random.randn(1000)

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color="orange", edgecolor='black') # 'bins' determines the number of bars

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Data")

# Show the plot
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, len(categories))

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")

# Show the plot
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (e.g., proportions of a whole)
sizes = [15, 30, 45, 10] # Proportions for each wedge
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4'] # Labels for each wedge
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] # Colors for each wedge
explode = (0.1, 0, 0, 0)  # "explode" the 1st slice (i.e. 'Category 1')

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Add title
plt.title("Simple Pie Chart")

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```
